```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
baseline = np.load("model_history/resnext_2_worker_128.npz", allow_pickle=True)

zero_worker = np.load("model_history/resnext_0_worker.npz", allow_pickle=True)
one_worker = np.load("model_history/resnext_1_worker.npz", allow_pickle=True)
four_worker = np.load("model_history/resnext_4_worker.npz", allow_pickle=True)
eight_worker = np.load("model_history/resnext_8_worker.npz", allow_pickle=True)

half_precision = np.load("model_history/resnext_2_worker_fp16.npz", allow_pickle=True)

baseline_large = np.load("model_history/resnext_2_worker_512.npz", allow_pickle=True)
checkpoint = np.load("model_history/resnext_2_worker_128_checkpoint.npz", allow_pickle=True)
checkpoint_large = np.load("model_history/resnext_2_worker_896_checkpoint.npz", allow_pickle=True)
```

# Data load

```python
plt.figure(figsize=(6, 5))
plt.plot([i for i in range(200)], baseline['loss'], 'r', label="baseline")
plt.plot([i for i in range(200)], zero_worker['loss'], 'g', label="zero_worker")
plt.plot([i for i in range(200)], one_worker['loss'], 'c', label="one_worker")
plt.plot([i for i in range(200)], four_worker['loss'], 'b', label="four_worker")
plt.plot([i for i in range(200)], eight_worker['loss'], 'y', label="eight_worker")
plt.legend(loc="upper right")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("loss with different number of worker")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
plt.plot([i for i in range(200)], baseline['acc'], 'r', label="baseline")
plt.plot([i for i in range(200)], zero_worker['acc'], 'g', label="zero_worker")
plt.plot([i for i in range(200)], one_worker['acc'], 'c', label="one_worker")
plt.plot([i for i in range(200)], four_worker['acc'], 'b', label="four_worker")
plt.plot([i for i in range(200)], eight_worker['acc'], 'y', label="eight_worker")
plt.legend(loc="lower right")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.title("training accuracy with different number of worker")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
plt.plot([i for i in range(200)], baseline['test_acc'], 'r', label="baseline")
plt.plot([i for i in range(200)], zero_worker['test_acc'], 'g', label="zero_worker")
plt.plot([i for i in range(200)], one_worker['test_acc'], 'c', label="one_worker")
plt.plot([i for i in range(200)], four_worker['test_acc'], 'b', label="four_worker")
plt.plot([i for i in range(200)], eight_worker['test_acc'], 'y', label="eight_worker")
plt.legend(loc="upper right")
plt.xlabel("epoch")
plt.ylabel("test accuracy")
plt.title("test accuracy with different number of worker")
plt.show()
```

# Precision

```python
plt.figure(figsize=(6, 5))
plt.plot([i for i in range(200)], baseline['loss'], 'r', label="baseline")
plt.plot([i for i in range(200)], half_precision['loss'], 'g', label="half_precision")
plt.legend(loc="upper right")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("loss with different precision")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
plt.plot([i for i in range(200)], baseline['acc'], 'r', label="baseline")
plt.plot([i for i in range(200)], half_precision['acc'], 'g', label="half_precision")
plt.legend(loc="lower right")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.title("training accuracy with different precision")
plt.show()
```

# Checkpoint

```python
plt.figure(figsize=(6, 5))
plt.plot([i for i in range(200)], baseline['loss'], 'r', label="baseline")
plt.plot([i for i in range(200)], baseline_large['loss'], 'g', label="baseline_large")
plt.plot([i for i in range(200)], checkpoint['loss'], 'c', label="checkpoint")
plt.plot([i for i in range(200)], checkpoint_large['loss'], 'b', label="checkpoint_large")
plt.legend(loc="upper right")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("loss with different batch size and checkpoint")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
plt.plot([i for i in range(200)], baseline['acc'], 'r', label="baseline")
plt.plot([i for i in range(200)], baseline_large['acc'], 'g', label="baseline_large")
plt.plot([i for i in range(200)], checkpoint['acc'], 'c', label="checkpoint")
plt.plot([i for i in range(200)], checkpoint_large['acc'], 'b', label="checkpoint_large")
plt.legend(loc="lower right")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.title("training accuracy with different batch size and checkpoint")
plt.show()
```
